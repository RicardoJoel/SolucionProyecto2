```python
#definimos la función que recodifica la variable “país de origen”
def recodificar_pais(fila):
  """devuelve recodificada la variable 'país de origen'"""
  #comparamos el pais con cada posible valor
  if fila["Pais_Origen"] == 'USA': 
    return 1
  elif fila["Pais_Origen"] == 'Europe': 
    return 2
  elif fila["Pais_Origen"] == 'Japan':
    return 3
```

```python
#usamos la función creada para recodificar la variable en el dataframe
data_autos["Pais_Origen"] = data_autos.apply(lambda x: recodificar_pais(x), axis=1)
data_autos
```

```python
#levantamos los datos con pandas y el delimitador ;
data_autos = pd.read_csv('autos.csv', delimiter=';')
data_autos.head()
```

```python
#obtenemos información del dataframe
data_autos.info()
```

```python
#evidenciamos la existencia de valores NULOS en las columnas "Millas_Por_Galon" y "Caballo_De_Fuerza"
data_autos.isnull().sum()
```

```python
#eliminamos las filas con Millas_Por_Galon o Caballo_De_Fuerza NULOS 
#debido a que serán usados posteriormente para el cálculo de nuevas variables
data_autos.dropna(subset=['Millas_Por_Galon','Caballo_De_Fuerza'], inplace=True)
```

```python
#verificamos la NO existencia de valores NULOS en el dataframe
data_autos.isnull().sum()
```

```python
#usamos Millas_Por_Galon y Cilindros para generar las 5 nuevas variables
data_autos["Nueva_1"] = np.log10(data_autos["Millas_Por_Galon"] / data_autos["Cilindros"])
data_autos["Nueva_2"] = np.sqrt(data_autos["Millas_Por_Galon"]) * np.exp(data_autos["Cilindros"]) / 200
data_autos["Nueva_3"] = np.where(data_autos["Millas_Por_Galon"] > data_autos["Cilindros"], 5, 3)
data_autos["Nueva_4"] = 1 / np.log(data_autos["Millas_Por_Galon"] / data_autos["Cilindros"]) * 100
data_autos["Nueva_5"] = data_autos["Cilindros"] ** 2 / data_autos["Millas_Por_Galon"]
```

```python
#verificamos la existencia de nuevas columnas
nuevas_columnas = ["Nueva_1", "Nueva_2", "Nueva_3", "Nueva_4", "Nueva_5"]
data_autos[nuevas_columnas]
```

```python
#definimos la función que compara cada variable con su media y desviación estandar
def evaluar(x, mean, std):
  if x > mean + 1 * std: 
    return 'Alto'
  elif x < mean - 1 * std: 
    return 'Bajo'
  else: #[med1 - 1 * std1 < x ; x < mean + 1 * std]
    return 'Medio'
```

```python
#con las variables previamente calculadas
var1 = data_autos["Nueva_1"]
var2 = data_autos["Nueva_2"]
var3 = data_autos["Nueva_3"]
#generamos las nuevas variables cualitativas ordinales
data_autos["Ordinal_1"] = var1.apply(lambda x: evaluar(x, var1.mean(), var1.std()))
data_autos["Ordinal_2"] = var2.apply(lambda x: evaluar(x, var2.mean(), var2.std()))
data_autos["Ordinal_3"] = var3.apply(lambda x: evaluar(x, var3.mean(), var3.std()))
```

```python
#verificamos la existencia de nuevas columnas
columnas_final = nuevas_columnas + ["Ordinal_1", "Ordinal_2", "Ordinal_3"]
data_autos[columnas_final]
```

```python
#procedemos a obtener las variables de tipo numérico
columnas = data_autos.select_dtypes(include=['float64', 'int']).columns
columnas
```

```python
#importamos las librerías necesarias para generar gráficos
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
#definimos la función que graficará cajas y bigotes por cada variable numérica
def graficar_cajas_bigotes(columnas, data, var):
  for i, colum in enumerate(columnas): 
    fig, ax = plt.subplots(figsize=(6, 3.84))
    sns.boxplot(x=data[var], y=colum, data=data)
    #damos formato al gráfico
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel(var)
```

```python
#se graficará según la variable cualitativa pasada como parámetro, p.e. Ordinal_1
graficar_cajas_bigotes(columnas, data_autos, "Ordinal_1")
```

```python
#definimos la función que graficará barras de media por cada variable numérica
def graficar_barras_media(columnas, data, var):
  for i, colum in enumerate(columnas):
    plt.figure()
    ax = data.iloc[i].plot(kind='bar')
    #damos formato al gráfico
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel(var)
```

```python
#obtenemos la media de cada variable numérica agrupada según el pais de origen
data_media = data_autos.groupby("Pais_Origen", sort=False).mean().T
data_media
```

```python
#se graficará según la variable cualitativa pasada como parámetro, p.e. Pais_Origen
graficar_barras_media(columnas, data_media, "Pais_Origen")
```

```python
#definimos la función que graficará histogramas con curva por cada variable numérica
def graficar_histo_curva(columnas, data, var):
  for i, colum in enumerate(columnas):
    ax = data.iloc[i].plot()
    #damos formato al gráfico
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel(var)
    #mostramos el gráfico
    plt.show()
```

```python
#obtenemos la media de cada variable numérica agrupada según el año
data_hist = data_autos.groupby("Anyo", sort=False).mean().T
data_hist
```

```python
#se graficará según la variable cualitativa pasada como parámetro, p.e. Anyo
graficar_histo_curva(columnas, data_hist, "Anyo")
```

```python
#consolidamos todo en una función que muestre gráficas por cada variable numérica
def mostrar_graficos(columnas, data, var):
  #la variable cualitativa es pasada como parámetro
  graficar_cajas_bigotes(columnas, data, var)
  media = data_autos.groupby(var, sort=False).mean().T
  graficar_barras_media(columnas, media, var)
  hist = data_autos.groupby(var, sort=False).mean().T
  graficar_histo_curva(columnas, hist, var)
```

```python
#probamos la función con el parámetro Anyo
mostrar_graficos(columnas, data_autos, "Anyo")
```

```python
#definimos la función de devuelve el mapa de calor con la correlación de variables numéricas
def mostrar_mapa_calor_correlacion(columnas, data):
  # para cada valor en la lista de posibles países
  for i, pais in enumerate(data["Pais_Origen"].unique()):
    # generamos el gráfico de correlación 
    correlacion = data[data["Pais_Origen"] == pais][columnas].corr()
    # obtenemos la máscara para el triángulo superior del gráfico
    mask = np.triu(np.ones_like(correlacion, dtype=bool)) 
    # la base para el gráfico
    f, ax = plt.subplots(figsize = (10,10))
    # colocamos un título
    ax.set_title(pais)
    # y lo representamos en un mapa de calor
    sns.heatmap(correlacion, 
                mask=mask, 
                cmap='Reds', vmax=1, center=0, 
                square=True, linewidths=.1, annot=True)
    
```

```python
#hacemos la llamada a la función con los parametros apropiados
mostrar_mapa_calor_correlacion(columnas, data_autos)
```
